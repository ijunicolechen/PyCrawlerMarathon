```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for weather in area:
    areaName = weather["locationName"]
    time = weather['weatherElement'][0]['time'][0]['dataTime']
    temperature = weather['weatherElement'][0]['time'][0]['elementValue']['value']
    print("地區：{}, 時間：{}, 溫度{} c".format(areaName, time, temperature))
```
